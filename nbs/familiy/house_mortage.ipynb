计算房贷等贷款信息
先从数据库提取出房贷信息，各年的房贷都是`house_debts`
注意`NULL`类型的存在，须将其过滤（Python提取后变成`None`类型）
首先输出一下各个房贷的分布

```python
range2010 = get('SELECT DISTINCT house_debts FROM famecon_2010 ORDER BY house_debts ASC')
range2018 = get('SELECT DISTINCT house_debts FROM famecon_2018 ORDER BY house_debts ASC')
print([num[0] for num in range2010])
print([num[0] for num in range2018])
```

获取一个比例，利用`COUNT`获取指定区间内的人数

```python
house_debt_2010 = []

house_debt_2010.append(count('famecon_2010', 'house_debts=0'))
house_debt_2010.append(count('famecon_2010', 'house_debts>0 and house_debts<=10000'))
house_debt_2010.append(count('famecon_2010', 'house_debts>10000 and house_debts<=100000'))
house_debt_2010.append(count('famecon_2010', 'house_debts>100000 and house_debts<=1000000'))
house_debt_2010.append(count('famecon_2010', 'house_debts>1000000'))

print(house_debt_2010)
```
